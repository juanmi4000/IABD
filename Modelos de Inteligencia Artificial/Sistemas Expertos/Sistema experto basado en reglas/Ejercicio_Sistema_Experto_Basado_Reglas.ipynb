```python
!pip install experta
!pip install --upgrade frozendict

from experta import *

class DiagnosticoPC(KnowledgeEngine):

  # Regla para diagnosticar conexión
  @Rule(Fact(no_enciende=True), Fact(sin_luces=True))
  def diagnostico_gripe(self):
      print("Diagnóstico: Revisar conexión eléctrica y fuente de alimentación.")


  @Rule(Fact(no_enciende=True), Fact(ventilador_funciona=True))
  def regla2(self):
      print("Diagnóstico: Verificar placa base o memoria RAM.")

  @Rule(Fact(sin_imagen=True), Fact(enciende=True))
  def regla3(self):
      print("Diagnóstico: Revisar cable de video.")

  @Rule(Fact(placa_base_enciende=True), Fact(sin_imagen=True))
  def regla3(self):
      print("Diagnóstico: Revisar tarjeta gráfica.")

  @Rule(Fact(funciona_lento=True))
  def regla4(self):
      print("Diagnóstico: Revisar programas en segundo plano o realizar análisis de virus.")

  @Rule(Fact(error_al_arrancar=True))
  def regla5(self):
      print("Diagnóstico: Verificar disco duro o reparar el sistema operativo.")

  @Rule(Fact(sin_sonido=True))
  def regla6(self):
      print("Diagnóstico: Revisar drivers de audio y conexiones de altavoces.")

  @Rule(Fact(wifi_no_funciona=True))
  def regla7(self):
      print("Diagnóstico: Verificar adaptador de red, reiniciar router o actualizar drivers.")

  @Rule(Fact(teclado_no_responde=True))
  def regla8(self):
      print("Diagnóstico: Probar otro puerto USB o cambiar teclado.")

  @Rule(Fact(raton_para=True))
  def regla9(self):
      print("Diagnóstico: Revisar si el puerto USB tiene polvo.")

  @Rule(Fact(raton_no_funciona=True))
  def regla9(self):
      print("Diagnóstico: Revisar batería (si es inalámbrico) o cambiar puerto USB.")

  @Rule(Fact(pantalla_azul=True))
  def regla10(self):
      print("Diagnóstico: Actualizar drivers o revisar hardware defectuoso.")

  @Rule(Fact(apagados_inesperados=True))
  def regla11(self):
      print("Diagnóstico: Revisar temperatura del procesador y ventilación.")

  @Rule(Fact(impresora_no_imprime=True))
  def regla12(self):
      print("Diagnóstico: Verificar conexión, tinta o drivers de la impresora.")

  @Rule(Fact(internet_lento=True))
  def regla13(self):
      print("Diagnóstico: Revisar velocidad de conexión o reiniciar router.")

  @Rule(Fact(teclado_para=True))
  def regla9(self):
      print("Diagnóstico: Revisar si el puerto USB tiene polvo.")

  @Rule(Fact(error_instalar_programa=True))
  def regla14(self):
      print("Diagnóstico: Verificar compatibilidad y permisos de instalación.")

  @Rule(Fact(usb_no_detectado=True))
  def regla15(self):
      print("Diagnóstico: Probar otro puerto o revisar drivers de USB.")

  @Rule(Fact(ip_incorrecta=True))
  def regla16(self):
      print("Diagnóstico: Revisa la IP.")

  @Rule(Fact(dns_error=True))
  def regla17(self):
      print("Diagnóstico: Comprueba que tenga la configuración del DNS.")

  @Rule(Fact(cortafuegos_bloqueo=True))
  def regla18(self):
      print("Diagnóstico: No tienes permisos para conectarte. Comprueba si tienes abierto el puerto.")



motor = DiagnosticoPC()
motor.reset()

print("Ingrese los síntomas del pc (Responda con 'si' o 'no'):")

sintomas = {
  "no_enciende": input("¿El PC no enciende? (si/no): ").strip().lower() == "si",
  "sin_luces": input("¿No hay luces en la torre? (si/no): ").strip().lower() == "si",
  "ventilador_funciona": input("¿El ventilador funciona? (si/no): ").strip().lower() == "si",
  "sin_imagen": input("¿No hay imagen en la pantalla? (si/no): ").strip().lower() == "si",
  "enciende": input("¿El PC enciende pero hay otro problema? (si/no): ").strip().lower() == "si",
  "placa_base_enciende": input("¿La placa base enciende pero hay otro problema? (si/no): ").strip().lower() == "si",
  "funciona_lento": input("¿Funciona  lento? (si/no): ").strip().lower() == "si",
  "error_al_arrancar": input("¿Has notado que te de errores al arrancar? (si/no): ").strip().lower() == "si",
  "sin_sonido": input("¿Te funciona el sonido? (si/no): ").strip().lower() == "no",
  "wifi_no_funciona": input("¿Otros dispositivos se conectan al wifi? (si/no): ").strip().lower() == "no",
  "teclado_no_responde": input("¿El teclado te responde? (si/no): ").strip().lower() == "no",
  "raton_para": input("¿El ratón se para? (si/no): ").strip().lower() == "si",
  "teclado_para": input("¿El teclado se para? (si/no): ").strip().lower() == "si",
  "raton_no_funciona": input("¿El ratón te funciona? (si/no): ").strip().lower() == "no",
  "pantalla_azul": input("¿Te han dado pantallazos azules? (si/no): ").strip().lower() == "si",
  "apagados_inesperados": input("¿Te han surgido apagados inesparados? (si/no): ").strip().lower() == "si",
  "impresora_no_imprime": input("¿La impresora funciona? (si/no): ").strip().lower() == "no",
  "internet_lento": input("¿El internet te va lento? (si/no): ").strip().lower() == "si",
  "error_instalar_programa": input("¿Te han dado errores al instalar el programa? (si/no): ").strip().lower() == "si",
  "usb_no_detectado": input("¿Has comprobado que esté conectado al USB? (si/no): ").strip().lower() == "si",
  "ip_incorrecta": input("¿Te ha dado problemas la IP? (si/no): ").strip().lower() == "si",
  "dns_error": input("¿La DNS te está dando proglemas? (si/no): ").strip().lower() == "si",
  "cortafuegos_bloqueo": input("¿El cortafuegos te está dando problemas para conectarte? (si/no): ").strip().lower() == "si"
}

# Declarar hechos en el sistema experto
for sintoma, presente in sintomas.items():
    if presente:
        motor.declare(Fact(**{sintoma: True}))


# Ejecutar el motor de inferencia
print("\nAnalizando los síntomas...\n")
motor.run()si
```
