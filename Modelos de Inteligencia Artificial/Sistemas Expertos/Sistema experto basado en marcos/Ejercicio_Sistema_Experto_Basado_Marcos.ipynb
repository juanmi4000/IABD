```python
class Enfermedad:
    def __init__(self, nombre, sintomas, tratamiento):
        self.nombre = nombre
        self.sintomas = sintomas
        self.tratamiento = tratamiento

    def __str__(self):
        return f" Enfermedad: {self.nombre}\n S칤ntomas: {', '.join(self.sintomas)}\n Tratamiento: {self.tratamiento}"

# Creaci칩n de la Base de Conocimiento

gripe = Enfermedad("Gripe", ["fiebre", "tos", "dolor de garganta"], "Reposo y l칤quidos")
resfriado_comun = Enfermedad("Resfriado com칰n", ["fiebre", "tos", "dolor de cabeza", "dolor muscular"], "Beber mucho l칤quido y reposo.")
covid_19 = Enfermedad("Covid-19", ["fiebre", "tos", "dificultad al respirar", "perdida de olfato y gusto"], "Realizar prueba de ant칤genos.")
neumonia = Enfermedad("Neumon칤a", ["fiebre", "dificultad para respirar", "tos fuerte"], "Antibi칩ticos y hospitalizaci칩n si es grave")
migra침a = Enfermedad("Migra침a", ["dolor de cabeza", "n치useas", "sensibilidad a la luz"], "Analg칠sicos y descanso")
alergia = Enfermedad("Alergia", ["estornudos", "congesti칩n nasal", "ojos llorosos"], "Ponerse mascarilla y alejarse de lo que le produzca alergia.")

# Base de conocimientos estructurada como un MARCO
base_conocimiento = {
    "gripe": gripe,
    "resfriado_comun": resfriado_comun,
    "covid19": covid_19,
    "neumonia": neumonia,
    "migra침a": migra침a,
    "alergia": alergia
}

print("游댌 **SISTEMA EXPERTO DE DIAGN칍STICO DE ENFERMEDADES** 游댌\n")
print("游늷 Selecciona una enfermedad:")
for indice, enfermedad in enumerate(base_conocimiento.values()):
  print(f"{indice + 1}. {enfermedad.nombre}")

opcion = input("\nIngrese el n칰mero del problema: ").strip()

mostrado = False

for indice, enfermedad in enumerate(base_conocimiento.values()):
  if str(indice + 1) == opcion:
    print(enfermedad)
    mostrado = True

if not mostrado:
  print("No se reconoce esa opci칩n")

```
